```r
# to generate exact same set of random numbers
set.seed(777)
rnorm(3) # 0.489786218319898 -0.398541384065542 0.510836321690603
# same as 'rnorm(10, mean=0, sd=1)'

set.seed(10)
x = rnorm(100, mean=90, sd=3)
mean(x) # 89.5903531699334
var(x) # 7.97332499072424
sd(x) # 2.82370766736294
# same as 'sqrt(var(x))'

set.seed(50)
y = rnorm(100, mean=100, sd=.1)
mean(y) # 99.9886472853026
var(y) # 0.00985270532066419
sd(y) # 0.0992607944793119

cor(x, y) # -0.100259186855998

plot(x, y, xlab="this is x-axis", ylab="this is y-axis", main="plot!", col="blue")

# pdf("figure.pdf")
# dev.off()
```
