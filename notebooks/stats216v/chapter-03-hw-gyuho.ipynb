A hypercube with side length 1 in d dimensions is defined to be the set of points $(x_{1}, x_{2}, ..., x_{d})$ such that $0 ≤ x_{j} ≤ 1$ for all $j = 1, 2, ..., d$. The boundary of the hypercube is defined to be the set of all points such that there exists a $j$ for which $0 ≤ x_{j} ≤ .05$ or $.95 ≤ x_{j} ≤ 1$ (namely, the boundary is the set of all points that have at least one dimension in the most extreme $10%$ of possible values). What proportion of the points in a hypercube of dimension 50 are in the boundary? (hint: you may want to calculate the volume of the non-boundary region)

Please give your answer as a value between 0 and 1 with 3 significant digits. If you think the answer is $50.52%$, you should say $0.505$:

<span style="color:blue">Gyu-Ho's Answer: 0.995</span>
- <span style="color:blue">The volume of hypercube with 50-dimension and side length 1 is $1^{50} = 1$.</span>
- <span style="color:blue">The volume of hypercube interior is $0.9^{50} = 0.005$.</span>
- <span style="color:blue">Thus, the volume of boundary is $1 - 0.005 = 0.995$.</span>

###### Q2 (a). How would you explain the difference between supervised and unsupervised learning? Can you give an example for each case that describes the differences?

<span style="color:blue">
Gyu-Ho's Answer: <b>Supervised learning</b> has outcome, <i>or dependent</i>, <b>variable</b> to guide statistical learning process. It seeks to predict <b>response variable</b> from <b>input variables</b>. <b>Unsupervised learning</b> does not measure or predict the outcome: <b>only observe features</b>. It seeks to understand how data are organized or clustered. Linear regression method is supervised learning, that can fit a model based on training observations and predict the response of previously unseen test data. For instance, linear regression can predict someone's wage, based on education level and seniority. In contrast, unsupervised learning wants probability distribution of entire dataset, or groups data into clusters of similar features. For instance, in marketing data of hundreds of features (age, gender, income, etc.), unsupervised learning can group those individuals to their shared demographic features.
</span>

###### Q2 (b). How would you explain the difference between regression and classification? Can you give an example where a regression methods could be used for classification? An example where a classification method could not be used for regression.

<span style="color:blue">
Gyu-Ho's Answer: Regression predicts numerical values, <i>such as stock price</i>, while classification predicts qualitative values, <i>such as posivie or negative</i>. Regression method can be used to predict qualitative values of levels. For instance, each observation is a student's previous test scores. And expect to estimate a response variable, <i>letter grade in the final exame (A, B, C, D, F)</i>. Regression first predicts the *test score out of 100 scale*, which can be converted to <i>letter grades</i> with certain thresholds. Regression assigns a numberical value to each response variable, whereas logistic regression, <i>or classification</i>, assigns probability, <i>or discrete values</i>. A classification method can not be used when regression requires explit ordering in output classes. For instance, classification method for predicting <i>letter grades in final exams</i> cannot be used to regress <i>numerical test score values</i>.
</span>

###### Q3. You are a data science consultant! In each of the following cases decide whether you would suggest a flexible regression model or an inflexible one. Provide your reasons for your client as clearly as possible.


<br>
(a) In the study of breast cancer, scientist are trying to find the associated genes. The total number of genes in the study is 50000 and the number of patients is 120.

<span style="color:blue">
Gyu-Ho's Answer: <b>Inflexible regression</b> because the number of predictors $p$ is extremely large, <i>5000</i>, and the sample size $n$ is relatively small, <i>120</i>. Flexible model would decrease <i>bias</i>, but might overfit the data due to small sample size. Plus, the study tries to find the associated genes, rather than predicting on unseen test data. Thus, it is more about <b>inference</b> to understand relationship between genes and cancer.
</span>


<br>
(b) The ministry of education in a certain country wants to identify student who need extra help. They wish to design a system which estimate student performance in the final 8th grad math exam based on their math, science and history grades in the 7th grade. To do this they want to run a regression on the data consisting of the information of all the students who have graduated from the 8th level in the last 10 years.

<span style="color:blue">
Gyu-Ho's Answer: <b>Flexible regression</b>, because the number of predictors $p$ is relatively small, <i>3</i>, and the sample size $n$ is extremely large, <i>all students who have graduated from the 8th level in the last 10 years</i>. Inflexible model would decrease <i>variance</i> being more interpretable but <i>highly biased</i> thus less accurate. Moreover, the ministry of education is not interested in understanding the relationships between variables. Thus, there is no need to choose trade-offs between flexibility and interpretability.
</span>

<br>
(c) Kelly is a very hardworking chemistry student and she has run an experiment to find a mathematical expression that explains the speed of corrosion of iron according to the humidity and temperature of the envirnoment and the percentage of different elements in the alloy. Unfortunately the lab that she is working in was established in 1967 and the equipment has not been changed seen then. She is skeptical about the quality of her measurements of the speed of corrosion.

<span style="color:blue">
Gyu-Ho's Answer: <b>Inflexible regression</b>, because flexible regression would be fitting irreducible error terms too closely. When $σ^{2} = Var(ϵ)$ is high due to poor quality of old lab facilities, flexible method would overfit to those errors. 
</span>


<br>
(d) Kelly’s advisor won the chemistry nobel prize and used the prize money to outfit the lab with the most modern equipment. Kelly has run her experiments again with the new equipments and now she can trust her numbers. But her advisor believes that she should not expect that the real relationship be linear.

<span style="color:blue">
Gyu-Ho's Answer: <b>Flexible regression</b> to find the nonlinear relationships between variables. For instance, linear regression, which is a <b>inflexible method</b>, assumes the linear relationship <b>with high bias</b>. If the true relationship is not linear, the model won't be accurate.
</span>
